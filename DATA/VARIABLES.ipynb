```python
import pandas as ps
```

```python
df = ps.read_csv('03_final_clean.csv')
```

```python
# Genero los dataframe por a√±o 

df_2015 = df[df['Year'] == 2015]
df_2016 = df[df['Year'] == 2016]
df_2017 = df[df['Year'] == 2017]
df_2018 = df[df['Year'] == 2018]
df_2019 = df[df['Year'] == 2019]
df_2020 = df[df['Year'] == 2020]
df_2021 = df[df['Year'] == 2021]
df_2022 = df[df['Year'] == 2022]
df_2023 = df[df['Year'] == 2023]
df_2024 = df[df['Year'] == 2024]


# Genero los dataframe por regiones

df_western_europe = df[df['Standardized_Region'] == 'Western Europe']
df_northern_europe = df[df['Standardized_Region'] == 'Northern Europe']
df_northern_america = df[df['Standardized_Region'] == 'Northern America']
df_australia_newzealand = df[df['Standardized_Region'] == 'Australia and New Zealand']
df_western_asia = df[df['Standardized_Region'] == 'Western Asia']
df_central_america = df[df['Standardized_Region'] == 'Central America']
df_south_america = df[df['Standardized_Region'] == 'South America']
df_south_eastern_asia = df[df['Standardized_Region'] == 'South-eastern Asia']
df_eastern_europe = df[df['Standardized_Region'] == 'Eastern Europe']
df_caribbean = df[df['Standardized_Region'] == 'Caribbean']
```

```python
df_2015.to_csv('YEARS//df_2015')
df_2016.to_csv('YEARS//df_2016')
df_2017.to_csv('YEARS//df_2017')
df_2018.to_csv('YEARS//df_2018')
df_2019.to_csv('YEARS//df_2019')
df_2020.to_csv('YEARS//df_2020')
df_2021.to_csv('YEARS//df_2021')
df_2022.to_csv('YEARS//df_2022')
df_2023.to_csv('YEARS//df_2023')
df_2024.to_csv('YEARS//df_2024')


# Genero los dataframe por regiones

df_western_europe.to_csv('REGIONS/western_europe')
df_northern_europe.to_csv('REGIONS/northern_europe')
df_northern_america.to_csv('REGIONS/northern_america')
df_australia_newzealand.to_csv('REGIONS/australia_newzealand')
df_western_asia.to_csv('REGIONS/western_asia')
df_central_america.to_csv('REGIONS/central_america')
df_south_america.to_csv('REGIONS/south_america')
df_south_eastern_asia.to_csv('REGIONS/south_eastern_asia')
df_eastern_europe.to_csv('REGIONS/eastern_europe')
df_caribbean.to_csv('REGIONS/caribbean')
```
